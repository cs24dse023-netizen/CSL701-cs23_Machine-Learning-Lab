```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("Titanic-Dataset (3).csv")

print("Dataset Loaded Successfully!")
```

```python
print("First 10 Rows:")
print(df.head(10))
```

```python
print("Last 5 Rows:")
print(df.tail(5))
```

```python
print("Dataset Shape:")
print(df.shape)
```

```python
print("Dataset Information:")
df.info()
```

```python
print(df.describe())

print("\nAge Statistics")
print("Mean:", df["Age"].mean())
print("Minimum:", df["Age"].min())
print("Maximum:", df["Age"].max())

print("\nFare Statistics")
print("Mean:", df["Fare"].mean())
print("Minimum:", df["Fare"].min())
print("Maximum:", df["Fare"].max())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
df["Age"].fillna(df["Age"].median(), inplace=True)

print("Missing Age Values After Filling:")
print(df["Age"].isnull().sum())
```

```python
survivors = df[df["Survived"] == 1]

print("Number of Survivors:")
print(survivors.shape[0])
```

```python
fare_array = df["Fare"].to_numpy()

print(fare_array)
```

```python
print("Maximum Fare:", np.max(fare_array))
print("Average Fare:", np.mean(fare_array))
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Survived", data=df)

plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Fare"], color="red", alpha=0.6)

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
